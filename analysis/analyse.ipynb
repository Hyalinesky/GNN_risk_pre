```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_csv('data/node.csv')  # 请将 'your_data.csv' 替换为你的数据文件路径

# 计算特征之间的相关性矩阵
correlation_matrix = data.corr()

# 绘制相关性热力图
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_csv('data/node.csv')  # 请将 'your_data.csv' 替换为你的数据文件路径

# 取部分特征或进行特征选择（这里只是示例，你可以根据需求调整选择的特征）
selected_features = data.iloc[:, :30]  # 选择前10个特征进行演示

# 计算特征之间的相关性矩阵
correlation_matrix = selected_features.corr()

# 绘制简化的相关性热力图
plt.figure(figsize=(8, 6))  # 调整图表大小
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Simplified Correlation Heatmap')
plt.show()
```
